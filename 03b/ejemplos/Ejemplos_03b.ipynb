```python
#  Desenfoque gaussiano y de caja

import cv2 as cv
import numpy as np

image = cv.imread('../images/Bosque.jpg')

cv.imshow('Original', image)

# utilizando filter2D

kernel = np.ones((11,1),np.float32)/11
dst = cv.filter2D(image,-1,kernel)
cv.imshow('Filter2D', dst)

# Suavizado por Promediado
avg = cv.blur(image, (11,11))
cv.imshow('AVG', avg)

# Suavizado gaussiano
image = cv.imread('../images/Bosque.jpg')

gaus = cv.GaussianBlur(image, (11,11), 100)
cv.imshow('Gaussian', gaus)


cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# C치lculo derivadas
import cv2 as cv
import numpy as np



image = cv.imread('images/Catedral.png')

cv.imshow('Color', image)

# Derivada en x
kernel = np.zeros((3,1),np.float32)
kernel[0] = -0.5
kernel[2] = 0.5
print(kernel)
Dfy = cv.filter2D(image.astype(np.float32),-1,kernel)
print(np.min(Dfy))

cv.imshow('Derivada y', Dfy.astype(np.uint8)+128)

# Derivada en x
kernel = np.zeros((1,3),np.float32)
kernel[0,0] = -0.5
kernel[0,2] = 0.5
print(kernel)
Dfx = cv.filter2D(image.astype(np.float32),-1,kernel)

cv.imshow('Derivada x', Dfx.astype(np.uint8)+128)

norm = np.sqrt(Dfx*Dfx+Dfy*Dfy)
norm[norm>255]=255
norm[norm<0]=0
cv.imshow('Norma', norm.astype(np.uint8))

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# C치lculo derivadas Sobel y Laplaciano
import cv2 as cv
import numpy as np

def normalizar(image):
    maximo = np.max(image)
    minimo = np.min(image)
    dif = maximo-minimo
    image_norm = 255*(image.astype(np.float32)-minimo)/dif
    return image_norm
    
image = cv.imread('images/Einstein.png')
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

cv.imshow('Original', gray)
gray = gray.astype(np.float32)

laplacian = cv.Laplacian(gray, -1)
dx = cv.Sobel(gray, -1, 1, 0)
dy = cv.Sobel(gray, -1, 0, 1)
print(np.max(dx), np.min(dx))
cv.imshow('Laplaciano',  normalizar(laplacian).astype(np.uint8))
cv.imshow('Derivada x',  normalizar(dx).astype(np.uint8))
cv.imshow('Derivada y',  normalizar(dy).astype(np.uint8))

norm = np.sqrt(dx*dx+dy*dy)
norm[norm>255]=255
norm[norm<0]=0
cv.imshow('Norma',  norm.astype(np.uint8))

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Filtrado mediano

import cv2 as cv
import numpy as np

image = cv.imread('images/noise/salt_pepper4.png')

#cv.imshow('Original', image)
median = cv.medianBlur(image,5)

# median blur
for i in range(100):
    median = cv.medianBlur(median,5)
    cv.imshow('Median', median)
    cv.waitKey(20)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Bilateral filtering

import cv2 as cv
import numpy as np

image = cv.imread('images/noise/gaussian_noise.jpg')
#image = cv.imread('images/noise/salt_pepper4.png')

cv.imshow('Original', image)

# bilateral blur
for i in range(50):
    bilateral = cv.bilateralFilter(image, 10, 65, 50)
    image=bilateral
    cv.imshow('Bilateral', image)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# Comparaci칩n entre filtros de suavizado ("blur")

import cv2 as cv
import numpy as np

image = cv.imread('images/noise/gaussian_noise.jpg')
#image = cv.imread('images/noise/salt_pepper4.png')

cv.imshow('Original', image)

#median blur
median=image.copy()
for i in range(5):
    median = cv.medianBlur(median,5)
cv.imshow('Median', median)
    
# bilateral blur
bilateral=image.copy()
for i in range(5):
    bilateral = cv.bilateralFilter(bilateral, 10, 65, 50)
cv.imshow('Bilateral', bilateral)

gaussian=image.copy()
for i in range(5):
    gaussian = cv.GaussianBlur(gaussian, (11,11), 0)
cv.imshow('Gaussian', gaussian)

cv.waitKey(0)
cv.destroyAllWindows()
```
